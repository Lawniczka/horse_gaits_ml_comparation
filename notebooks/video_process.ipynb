```python
import cv2
import os

horse = "../dataset/dataset_20210201 Charles walk_36_1612216820383.avi"
dir = "../dataset/"
out = "../output/learn/vid/"

os.makedirs(out, exist_ok=True)
```

```python
cap = cv2.VideoCapture(horse)

frame_count = 0
save_every = 10

if not cap.isOpened():
    print("Nie można otworzyć wideo")
    exit()
    
while True:
    ret, frame = cap.read()
    
    if not ret:
        break
    
    frame_count += 1
    
    cv2.imshow("Klatka", frame)
    
    if frame_count % save_every == 0:
        cv2.imwrite(f"{out}/frame_{frame_count}.png",frame)
        
    if cv2.waitKey(1) & 0xff == ord('q') or cv2.getWindowProperty("Klatka", cv2.WND_PROP_VISIBLE) < 1:
        break
    
cap.release()
cv2.destroyAllWindows()
```
