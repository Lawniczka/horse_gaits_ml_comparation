```python
import cv2
import numpy as np
```

```python
sample = 1
label = "walk"
edges_dir = "../../dataset/preprocessed/edges/" + label + '/' + label + str(sample).zfill(4) + "/"

edges_dir = "../dataset/preprocessed/edges/" + label + "/0001/"

T = 32
```

# Variance before resize

```python
sum_ = np.zeros((256,256), dtype=np.float32)
sum_sq = np.zeros((256,256), dtype=np.float32)

for i in range(T):
    frame = np.load(edges_dir + f"{str(i+1).zfill(7)}.npy").astype(np.float32)
    sum_ += frame
    sum_sq += frame ** 2
    
mean = sum_ / T
var = (sum_sq / T) - (mean ** 2)

before = cv2.resize(var, (128, 128), interpolation=cv2.INTER_AREA)

before = cv2.normalize(before, None, 0, 255, cv2.NORM_MINMAX).astype(np.uint8)
```

# Variance after resize

```python
sum_ = np.zeros((128,128), dtype=np.float32)
sum_sq = np.zeros((128,128), dtype=np.float32)

for i in range(T):
    frame = np.load(edges_dir + f"{str(i+1).zfill(7)}.npy").astype(np.float32)
    frame = cv2.resize(frame, (128, 128), interpolation=cv2.INTER_AREA)
    sum_ += frame
    sum_sq += frame ** 2
    
mean = sum_ / T
var = (sum_sq / T) - (mean ** 2)

after = cv2.normalize(var, None, 0, 255, cv2.NORM_MINMAX).astype(np.uint8)
```

# Compare

```python
import cv2
import numpy as np

print(np.mean(before), np.mean(after))
print(np.max(before) - np.max(after))
print(np.array_equal(before, after))
```

```python
frames = [np.load(f"{edges_dir}{str(i+1).zfill(7)}.npy") for i in range(32)]

frames_resized = [cv2.resize(f.astype(np.float32), (128,128), interpolation=cv2.INTER_AREA) for f in frames]
var_before = np.var(np.stack(frames_resized), axis=0)
var_before = cv2.normalize(var_before, None, 0, 255, cv2.NORM_MINMAX).astype(np.uint8)

frames_var = np.var(np.stack(frames), axis=0)
var_after = cv2.resize(frames_var, (128,128), interpolation=cv2.INTER_AREA)
var_after = cv2.normalize(var_after, None, 0, 255, cv2.NORM_MINMAX).astype(np.uint8)
```

```python
print(np.mean(var_before), np.mean(var_after))
print(np.max(var_before) - np.max(var_after))
print(np.array_equal(var_before, var_after))
```
